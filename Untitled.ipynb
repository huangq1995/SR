```python
filenames = [];
spatial_freq = [];
corr = [];
ctf_fit = [];
for filename in os.listdir('need_to_plot_4'):
    if filename.endswith(".txt"):
        filedir = 'need_to_plot_4'+'/'+filename
        filenames.append(filename)
        with open(filedir) as f:
            for i, j in enumerate(f):
                if i == 5:
                    j = j.strip()
                    j = j.split()
                    j = [float(i) for i in j]
                    spatial_freq.append(j)
                if i == 9:
                    j = j.strip()
                    j = j.split()
                    j = [float(i) for i in j]
                    corr.append(j)
                if i == 8:
                    j = j.strip()
                    j = j.split()
                    j = [float(i) for i in j]
                    ctf_fit.append(j)
        
```
